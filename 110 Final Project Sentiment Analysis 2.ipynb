```python
### I first imported the required libraries.
import json
import demoji
from csv import writer
from apiclient.discovery import build

### Next, I configured the function parameters for the required variables to pass to the service.
def get_comments(part='snippet', 
                 maxResults=100, 
                 textFormat='plainText',
                 order='time',
                 videoId='LyKuUxkDBf8',
                 csv_filename="SaskDutchKid_channel"):

    ### I now need to make some empty lists to store the information that I want: comment, commentID, and replyCount.
    comments, commentsId, repliesCount = [], [], []
       
    ### I then built the service for calling the YouTube API.
    def build_service(filename):
        key= 'AIzaSyBNKAPHQDx3nvGzE7RgIDOoKLs-Vr5YIDw'
        YOUTUBE_API_SERVICE_NAME = "youtube"
        YOUTUBE_API_VERSION = "v3"
        return build(YOUTUBE_API_SERVICE_NAME,
                 YOUTUBE_API_VERSION,
                 developerKey=key)
    ### I chose the first video and called the service.
    service = build_service('api_key.txt') 
    response_tp = service.commentThreads().list(
        part='snippet',
        maxResults=100,
        textFormat='plainText',
        order='time',
        videoId='LyKuUxkDBf8'
    ).execute()
    
    ### I have noticed that there are more comments than my available maximum of 100, 
    ### so I need to make a for loop to ensure I use all of the available comments.
      
    while response_tp: 
                 
        for item in response_tp['items']:
            ### This indexes the items for the three data features that I want.
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comment_id = item['snippet']['topLevelComment']['id']
            reply_count = item['snippet']['totalReplyCount']
            
            ### This must also append to the lists.
            comments.append(comment)
            commentsId.append(comment_id)
            repliesCount.append(reply_count)
            
            print(comment)
            print(comment_id)
            print(reply_count)
            
        
        try:
        ### I must also check for nextPageToken and set the response equal to the JSON response if it does exist.
            if 'nextPageToken' in response_tp:
                response_tp = service.commentThreads().list(
                    part=part,
                    maxResults=maxResults,
                    textFormat=textFormat,
                    order=order,
                    videoId=videoId,
                    pageToken=response_tp['nextPageToken']
            ).execute()
            else:
                break
        except: break   

    ### Finally, I want to return the data that I am interested in.
    return {
        'Comments': comments,
        'Comment ID': commentsId,
        'Reply Count' : repliesCount
    }
  
```

```python
df2= df1.dropna(subset=["clean_comments"])
```

```python
sentiment = SentimentIntensityAnalyzer()
sentiment.polarity_scores(df2.clean_comments[0])
df2['SDK_sentiment'] = df2.clean_comments.apply(lambda x: sentiment.polarity_scores(x))
```

```python
df2['SDK_neg_sentiment'] = df2.SDK_sentiment.apply(lambda x: x['neg'])
df2['SDK_pos_sentiment'] = df2.SDK_sentiment.apply(lambda x: x['pos'])
df2['SDK_comp_sentiment'] = df2.SDK_sentiment.apply(lambda x: x['compound'])
```
